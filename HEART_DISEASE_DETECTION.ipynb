```python
# As we will be using both types of approches for demonstration lets do First Label Ecoding
# which will be used with Tree Based Algorthms
df_tree = df.apply(LabelEncoder().fit_transform)
df_tree.head()
```

```python
## Creaeting one hot encoded features for working with non tree based algorithms
df_nontree=pd.get_dummies(df,columns=string_col,drop_first=False)
df_nontree.head()
```

```python
# Getting the target column at the end
target="HeartDisease"
y=df_nontree[target].values
df_nontree.drop("HeartDisease",axis=1,inplace=True)
df_nontree=pd.concat([df_nontree,df[target]],axis=1)
df_nontree.head()
```

USING LOGESTIC REGRESSION

```python
feature_col_nontree=df_nontree.columns.to_list()
feature_col_nontree.remove(target)
```

```python
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,roc_auc_score
from sklearn.preprocessing import RobustScaler,MinMaxScaler,StandardScaler
acc_log=[]

kf=model_selection.StratifiedKFold(n_splits=5)
for fold , (trn_,val_) in enumerate(kf.split(X=df_nontree,y=y)):

    X_train=df_nontree.loc[trn_,feature_col_nontree]
    y_train=df_nontree.loc[trn_,target]

    X_valid=df_nontree.loc[val_,feature_col_nontree]
    y_valid=df_nontree.loc[val_,target]

    #print(pd.DataFrame(X_valid).head())
    ro_scaler=MinMaxScaler()
    X_train=ro_scaler.fit_transform(X_train)
    X_valid=ro_scaler.transform(X_valid)


    clf=LogisticRegression()
    clf.fit(X_train,y_train)
    y_pred=clf.predict(X_valid)
    print(f"The fold is : {fold} : ")
    print(classification_report(y_valid,y_pred))
    acc=roc_auc_score(y_valid,y_pred)
    acc_log.append(acc)
    print(f"The accuracy for Fold {fold+1} : {acc}")
    pass
```

```python
import os
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from xgboost import XGBClassifier
import shap
```

```python
# Load the trained XGBoost model
model = XGBClassifier()
model.load_model("/content/heart_disease_model.bin")
```

```python
# Get user input
age = int(input("Enter your age: "))
sex = int(input("Enter your sex (1 for male, 0 for female): "))
cp = int(input("Enter your chest pain type (0 for typical angina, 1 for atypical angina, 2 for non-anginal pain, 3 for asymptomatic): "))
trestbps = int(input("Enter your resting blood pressure: "))
chol = int(input("Enter your cholesterol level: "))
fbs = int(input("Enter your fasting blood sugar level (1 if > 120 mg/dl, 0 otherwise): "))
restecg = int(input("Enter your resting ECG results (0 for normal, 1 for ST-T wave abnormality, 2 for left ventricular hypertrophy): "))
thalach = int(input("Enter your maximum heart rate: "))
exang = int(input("Enter your exercise-induced angina (1 for yes, 0 for no): "))
oldpeak = float(input("Enter your ST depression induced by exercise relative to rest: "))
slope = int(input("Enter the slope of the peak exercise ST segment (0 for upsloping, 1 for flat, 2 for downsloping): "))

# Create a DataFrame with the user input
user_input = pd.DataFrame({
    "Age": [age],
    "Sex": [sex],
    "ChestPainType": [cp],
    "RestingBP": [trestbps],
    "Cholesterol": [chol],
    "FastingBS": [fbs],
    "RestingECG": [restecg],
    "MaxHR": [thalach],
    "ExerciseAngina": [exang],
    "Oldpeak": [oldpeak],
    "ST_Slope": [slope],
})

# Encode categorical features
user_input_encoded = pd.get_dummies(user_input, columns=["ChestPainType", "RestingECG", "ST_Slope"])

# Scale the data
scaler = MinMaxScaler()
user_input_scaled = scaler.fit_transform(user_input_encoded)
```

```python
# Initialize the SHAP explainer with the trained model
explainer = shap.Explainer(model)

# Get SHAP values for the user input
shap_values = explainer.shap_values(user_input_scaled)

# Visualize the SHAP values
shap.initjs()
shap.force_plot(explainer.expected_value, shap_values, user_input_encoded)
```

```python
# Make a prediction using the trained model
prediction = model.predict(user_input_scaled)[0]

# Print the prediction
if prediction == 0:
    print("You are predicted to not have heart disease.")
else:
    print("You are predicted to have heart disease.")
```
